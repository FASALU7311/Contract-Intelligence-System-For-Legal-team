```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Load the dataset
df = pd.read_csv("cuad_final.csv")

# Display basic info
print(df.info())
print(df.head())

# Label distribution
plt.figure(figsize=(14,6))
df['clause_type'].value_counts().head(20).plot(kind='bar', color='black')
plt.title("Top 20 Clause Types by Frequency")
plt.xlabel("Clause Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Word count per clause
df['word_count'] = df['clause_text'].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(10,5))
sns.histplot(df['word_count'], bins=50, kde=True, color='red')
plt.title("Clause Length Distribution (in words)")
plt.xlabel("Number of Words")
plt.ylabel("Number of Clauses")
plt.show()

# Optional: dummy model evaluation from training logs
y_true = df['clause_type']
y_pred = df['clause_type']  # replace with actual predictions if available

print("Classification Report:")
print(classification_report(y_true, y_pred))  # for now, dummy same-as-true

# Dummy Confusion Matrix (again, replace with model output if available)
label_names = df['clause_type'].unique()
cm = confusion_matrix(y_true, y_pred, labels=label_names)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=False, fmt='d', xticklabels=label_names, yticklabels=label_names, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
